```python
def findname(namestring):
    name_match = re.search(r"\d*\s*(J*\d+\.*\d*\+*\-*\d+\.*\d*)", namestring)
    if name_match:
        name = name_match.group(1)
        name1= name[:5]
        name2= re.search(r"J*\d+\.*\d*(\+*\-*\d+\.*\d*)", name).group(1)[:5]
        name = name1+name2
    else:
        name = ''
    return name
```

```python
def classifbinaries(df):
    tm = np.logical_or(np.logical_and(wave > 1.35,wave < 1.42), np.logical_and(wave > 1.8,wave < 1.95))
    classif=[]
    classif_t=[]
    classif_ba=[]
    classif_bu=[]
    for i in range(len(df)):
        name = findname(df.name.iloc[i])   
        sp0= sp.getSpectrum(shortname = name)[0]    
        wave0  = sp0.wave.value
        flux0  = np.array(interpolate_flux_wave(wave0, sp0.flux.value, wgrid=wavegrid))
        noise0 = interpolate_flux_wave(wave0, sp0.noise.value, wgrid=wavegrid)
        flux0, noise0 = normalize(wave, flux0, noise0)
        fluxT = flux0[~tm]
        noiseT = noise0[~tm]
        SNR = measureSN(wave, flux0, noise0, rng=[1.2, 1.35])
        fastspt = typeToNum(fast_classify(wave,flux0,noise0))
        if SNR<50:
            classif.append(list(mo_low.predict([flux0])[0]))
            classif_t.append(list(mo_low_T.predict([fluxT])[0]))
            # if 16<=fastspt<=33:
            if 10<=fastspt<=50:
                classif_ba.append(list(mo_low_ba.predict([flux0])[0]))
            else:
                classif_ba.append([])
            # if 22<=fastspt<=34:
            if 10<=fastspt<=50:
                classif_bu.append(list(mo_low_bu.predict([flux0])[0]))
            else:
                classif_bu.append([])
        elif SNR>100:
            classif.append(list(mo_hig.predict([flux0])[0]))
            classif_t.append(list(mo_hig_T.predict([fluxT])[0]))
            # if 16<=fastspt<=33:
            if 10<=fastspt<=50:
                classif_ba.append(list(mo_mid_ba.predict([flux0])[0]))
            else:
                classif_ba.append([])
            # if 22<=fastspt<=34:
            if 10<=fastspt<=50:
                classif_bu.append(list(mo_mid_bu.predict([flux0])[0]))
            else:
                classif_bu.append([])
        else:
            classif.append(list(mo_mid.predict([flux0])[0]))
            classif_t.append(list(mo_mid_T.predict([fluxT])[0]))
            # if 16<=fastspt<=33:
            if 10<=fastspt<=50:
                classif_ba.append(list(mo_low_ba.predict([flux0])[0]))
            else:
                classif_ba.append([])
            # if 22<=fastspt<=34:
            if 10<=fastspt<=50:
                classif_bu.append(list(mo_low_bu.predict([flux0])[0]))
            else:
                classif_bu.append([])
    return classif, classif_t, classif_ba, classif_bu
```
